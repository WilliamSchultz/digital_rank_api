```python
import pandas as pd
import requests
```

```python
list_of_domains =[
    'hornbach.de',
    'bauhaus.info',
    'hagebau.de',
    'kaufland.de',
    'globus-baumarkt.de',
    'manomano.de',
    'mein-schoener-garten.de',
    'otto.de',
    'bauhaus.at',
    'hornbach.at',
]
```

```python
list_urls = [f'https://api.similarweb.com/v1/similar-rank/{domain}/rank?api_key={key}' for domain in list_of_domains]
```

```python
def get_data(r):
    string = r.json()
    domain = string['meta']['request']['domain']
    rank = string['similar_rank']['rank']
    return domain, rank
```

```python
results = []
for url in list_urls:
    r = requests.get(url)
    tup = get_data(r)
    results.append(tup)
```

```python
pd.DataFrame(results)
```
